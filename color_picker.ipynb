```python
from skimage import io
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
```

```python
img = io.imread("/Users/zhengyang/Desktop/Weixin Image_20260108145114_16_182.png")
```

```python
plt.imshow(img)
```

```python
img = img[300:400, :, :]
```

```python
mean = img.mean(axis=0, keepdims=True).astype(int)
smooth = gaussian_filter1d(mean, sigma=15, axis=1)
```

```python
plt.figure(dpi=300)
plt.imshow(img)
plt.axis("off")
```

```python
plt.figure(dpi=300)
plt.imshow(mean, aspect=20)
plt.axis("off")
# plt.colorbar()
```

```python
plt.figure(dpi=300)
plt.imshow(smooth, aspect=20)
plt.axis("off")
```
