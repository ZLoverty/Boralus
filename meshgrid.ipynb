```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 50)
X, Y = np.meshgrid(x, y)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(4, 4))
ax[0].imshow(X)
ax[1].imshow(Y)
```

```python
h = np.sin(X+Y)
```

```python
plt.imshow(h)
```
